# Drop unnamed column

```python
dataset = dataset.drop("Unnamed: 0", axis =1)
```

```python
dataset.head(5)
```

# Gender Distribution

```python
plt.figure(figsize = (4,4))
ax=sns.countplot(x="Gender", data=dataset, color="orange")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

from the above chart we have analysed that:
the number of females in the data is more than the number of males

# Relationship between parent's education and student's score

```python
gb = dataset.groupby("ParentEduc").agg({"MathScore": "mean", "ReadingScore":"mean","WritingScore":"mean"})
```

```python
gb
```

```python
plt.figure(figsize = (5,5))
sns.heatmap(gb, annot= True)
plt.title("Relationship between parent's education and student's score")
plt.show()
```

from the above chart we have concluded that the education of the parents have a good impact on their students

# Relationship between parent's Marital Status and student's score

```python
gb1 = dataset.groupby("ParentMaritalStatus").agg({"MathScore": "mean", "ReadingScore":"mean","WritingScore":"mean"})
gb1
```

```python
plt.figure(figsize = (5,5))
sns.heatmap(gb1, annot= True)
plt.title("Relationship between parent's Marital Status and student's score")
plt.show()
```

from the above chart we have concluded that there is no/negligible impact on the student's score due to their parent marital status

# Distribution of Ethnic Group

```python
groupA = dataset.loc[dataset["EthnicGroup"] == "group A"].count()
groupB = dataset.loc[dataset["EthnicGroup"] == "group B"].count()
groupC = dataset.loc[dataset["EthnicGroup"] == "group C"].count()
groupD = dataset.loc[dataset["EthnicGroup"] == "group D"].count()
groupE = dataset.loc[dataset["EthnicGroup"] == "group E"].count()

l = ["group A","group B","group C","group D","group E"]
mlist = [groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]

plt.pie(mlist, labels = l, autopct = "%0.2f%%")
plt.title("Distribution of Ethnic Groups")
plt.legend(l, loc='upper left')
plt.show()

```
